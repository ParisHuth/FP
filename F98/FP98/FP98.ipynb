```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.image as mpimg
%matplotlib inline
import numpy as np
from numpy import exp, sqrt, log, pi
from scipy.optimize import curve_fit
from scipy.stats import chi2
from scipy import odr
from scipy.integrate import quad
from scipy.signal import find_peaks
from scipy.signal import argrelextrema, argrelmin, argrelmax
from scipy.special import erfc
from sympy.solvers import solve
from sympy import Symbol
import pandas as pd
import matplotlib.transforms as mtransforms

def fehler(name, G, sig_G, G_lit, sig_G_lit):
    print(name)
    print('Relativer Fehler: ', sig_G / G * 100)
    print('Rel. Fehler (Vergleich):', sig_G_lit / G_lit * 100)
    print('Absoluter Fehler: ', G - G_lit)
    print('Verhältnis:', G / G_lit)
    print('Sigma-Abweichung: ', np.abs(G - G_lit) / sqrt(sig_G ** 2 
                                                         + sig_G_lit ** 2),'\n')
    
    
def fehler_small(name, G, sig_G):
    print(name)
    print('Relativer Fehler: ', sig_G / G * 100)
    

def ergebnis(name, G, sig_G, komma, einheit):
    print(name + ' =', np.round(G, komma), '+/-', np.round(sig_G, komma), einheit)
    

def ergebnis_large(name, G, sig_G, komma, einheit):
    print(name + ' =', np.round(G, komma))
    print('+/-'.rjust(len(name) + 2), np.round(sig_G, komma), einheit)
    

def fitparameter(name, G, sig_G, komma, einheit):
    return name + ' =' + str(np.round(G, komma)) + '$\pm$' + str(np.round(sig_G, komma)) + einheit

def chi_sq(test_func, x_val, y_val, sig_y_val, popt, number):
    chi2_ = np.sum((test_func(x_val, *popt) - y_val) ** 2 
                   / sig_y_val ** 2)
    dof = len(y_val) - number
    chi2_red = chi2_ / dof

    print("chi2 =", chi2_)
    print("chi2_red =",chi2_red)

    prob = np.round(1 - chi2.cdf(chi2_,dof), 2) * 100 
    print("Probability =", prob, "%\n")
```

### 5.2.1 Offener Regelkreis

#### Widerstand
Wir möchten den Widerstand des normalleitenden SQUID messen, also oberhalb der kritischen Temperatur, bei der die supraleitenden Eigenschaften nicht mehr auftreten. Dazu zeichnen wir die charakteristische Strom-Spannungs-Kurve auf. Es wurden mehrere Messwerte für die Spannung und die zugehörigen Stromwerte aufgenommen.

Für diese Messung wurde der Spannungswert mit einem Oszilloskop bestimmt, indem zu verschiedenen Stromwerten jeweils ein Zeitpunkt zur Messung ausgewählt wurde. Dadurch ergibt sich ein geringer menschlicher Messfehler, da der Zeitpunkt der Messung manuell festgelegt wurde. Zusätzlich können Spannungsschwankungen auftreten. Daher schätzen wir den maximalen Fehler großzügig auf etwa 5 mV.

Gemäß dem Ohmschen Gesetz ergibt sich der Zusammenhang zwischen Spannung und Strom durch

$$R = \frac{V}{I}.$$

Diese Gleichung wenden wir im linearen Bereich der Strom-Spannungs-Kennlinie an, um den Widerstand des SQUID zu bestimmen. Es ist dabei wichtig zu beachten, dass sich die Kennlinie im nicht-supraleitenden Zustand wie ein gewöhnlicher metallischer Widerstand verhält, was eine lineare Beziehung zwischen Spannung und Strom bedeutet. Abweichungen in nicht-linearen Bereichen würden auf andere Phänomene oder Messfehler hinweisen.

#### Strom-Spannungs-Kennlinie für ganzzahlige und halbzahliger Flüsse

Als Nächstes haben wir die Strom-Spannungs-Kennlinie für ganzzahlige und halbzahliger Vielfache des Flusses $\Phi_0$ gespeichert. Die Kennlinie für ganzzahlige Vielfache von $\Phi_0$ lässt sich durch ihr breiteres Plateau erkennen. Dies ist auch in Abbildung 1 und 2 zu erkennen.

In der SQUID-Physik führt das Anlegen eines magnetischen Flusses in ganzzahligen und halbzahliger Vielfachen von  $\Phi_0$ zu unterschiedlichen Eigenschaften der Strom-Spannungs-Kennlinie. Während für ganzzahlige Vielfache von $\Phi_0$ ein stabileres und breiteres Plateau in der Kennlinie entsteht, zeigt die Kurve bei halbzahliger Vielfachen oft eine stärkere Steigung und weniger ausgeprägte Plateaus. Diese Unterschiede resultieren aus der symmetrischen und asymmetrischen Flussquantisierung im SQUID und beeinflussen das Verhalten des Systems unter verschiedenen Magnetfeldern.

```python
# Function to read files
def readfiles(data):
    x = np.loadtxt('data/' + data, skiprows = 3, delimiter=';'
                   , converters={0: lambda s: float(s.replace(b',', b'.')),
                                1: lambda s: float(s.replace(b',', b'.')),
                                2: lambda s: float(s.replace(b',', b'.'))})
    
    return x

def readfiles2(data):
    x = np.loadtxt('data/' + data, skiprows = 3, delimiter=';'
                   , converters={0: lambda s: float(s.replace(b',', b'.')),
                                1: lambda s: float(s.replace(b',', b'.'))})
    
    return x

# Definiere die lineare Funktion für die Anpassung
def line(x, m, b):
    return m * x + b
```

#### Mutual Inductances
Wir können die gegenseitigen Induktivitäten $M_{\text{IN}}^{-1}$ und $M_{\Phi\text{B}}^{-1}$ mithilfe der Gleichung (47) im Skript berechnen:

$$ M^{-1} = \dfrac{I_\Phi}{\Delta \Phi} $$

```python
# Mutual inductance für X

delta_Phi_X = 5 # Phi_0
err_Phi_X = 0.1
tplus_X = np.argmax(x_521_X[:, 1])
tminus_X = np.argmin(x_521_X[:, 1])
t_max_X = x_521_X[tplus_X, 0] - x_521_X[tminus_X, 0]

I_Phi_X = abs(43.04 * (x_521_X[time2_X, 0] - x_521_X[time1_X, 0]) / t_max_X)
# 43.04 *: generator signal, time_2-1 / t_max: Anteil an Strom was wir betrachten

M_Phi_X_inv = (delta_Phi_X / I_Phi_X) ** -1
err_M_Phi_X_inv = I_Phi_X * (err_Phi_X) / (delta_Phi_X ** 2)


# Mutual inductance für B
delta_Phi_B = 1.0 # Phi_0
err_Phi_B = 0.1
tplus_B = np.argmax(x_521_B[:, 1])
tminus_B = np.argmin(x_521_B[:, 1])
t_max_B = x_521_B[tplus_B, 0] - x_521_X[tminus_B, 0]

I_Phi_B = abs(43.06 * (x_521_B[time2_B, 0] - x_521_X[time1_B, 0]) / t_max_B) # uA
err_I_Phi_B = 0


M_Phi_B_inv = (delta_Phi_B / I_Phi_B) ** -1
err_M_Phi_B_inv = I_Phi_B * (err_Phi_B) / (delta_Phi_B ** 2)
```

Die Differenz $\Delta \Phi$ wird durch die Anzahl der Flussquanten (Perioden) in einem Zyklus bestimmt. Um die Fehler zu ermitteln, müssen wir Folgendes berücksichtigen: a) den „Lese“-Fehler der Flussquanten, d. h. ob wir einen Peak gezählt haben, der nicht zum Zyklus gehört. Wir gehen von einem Lese-Fehler von $\pm 0.1$ aus, obwohl die Werte theoretisch diskret in natürlichen Zahlen vorliegen würden. Kleine Fluktuationen sind jedoch innerhalb dieses Fehlers berücksichtigt.

Das Generatorsignal hatte im Verhältnis zu den „Zählungen“ eine sehr hohe Genauigkeit, sodass dessen Einfluss auf den Fehler vernachlässigbar ist.

Für $M^{-1}_{\Phi\text{B}}$ mit

```python
ergebnis('Phi_B', delta_Phi_B, err_Phi_B, 1, 'Phi_0')
```

mit einem Fehler von

$$ \Delta M^{-1}_{\Phi\text{B}} = I_{\Phi_B} \cdot \dfrac{\Delta \Phi_B}{\Phi_B^2}$$

bekommen wir

Damit erhalten wir einen Verstärkungsfaktor von

```python
ergebnis('m_exp', m_exp, err_m_exp, 5, 'V uA^-1')
```

```python
# Theoretische Verstärkung basierend auf den Werten für die gegenseitigen Induktivitäten
R_F = 100 # kOhm
err_m_theo = R_F * sqrt( (err_M_Phi_B_inv / M_Phi_X_inv) ** 2 
                   + (M_Phi_B_inv * err_M_Phi_X_inv / (M_Phi_B_inv) ** 2) ** 2 ) * 1e-3
m_theo = R_F * M_Phi_B_inv / M_Phi_X_inv * 1e-3
```

Der Verstärkungsfaktor kann durch die folgende Gleichung berechnet werden:
    
$$ \Delta U_{out} = - R_F \dfrac{M_{\text{IN}}}{M_{\Phi\text{B}}} \Delta I_{\text{IN}} $$ 

wobei $M_\text{IN}^{-1} = M_{\Phi X}^{-1}$ gilt. Der Verstärkungsfaktor ergibt sich dann als 

$$ m = R_F \dfrac{M_{\text{IN}}}{M_{\Phi \text{B}}}. $$

Für diese Messung berücksichtigen wir einen Widerstandsfehler von $\pm 1\%$, $R_F = 100(1)\, \text{k}\Omega$. Im Vergleich zu den Fehlern der gegenseitigen Induktivitäten ist dieser Fehler jedoch vernachlässigbar. Der Gesamtfehler des Verstärkungsfaktors berechnet sich zu

$$ \Delta m = R_F \sqrt{(\dfrac{\Delta M_{{\text{IN}}^{-1}}}{M_{\Phi\text{B}}^{-1}})^2 + ( \dfrac{\Delta M_{\text{IN}}^{-1} M_{\Phi\text{B}}}{(M_{\text{IN}^{-1}})^2})^2 } $$ 

Verwendet man die Werte aus der vorherigen Aufgabe für [$M^{-1}_{\text{IN}}$](#M_Phi_X_inv) und [$M^{-1}_{\Phi B}$](#M_Phi_B_inv) erhalten wir

```python
ergebnis('Theoretical amplification: m_theo', m_theo, err_m_theo, 3, '[V microA^-1]')
```

Die Abweichung zwischen den beiden Werten beträgt

```python
fehler('Amplification: experimental and theoretical (comparison)', m_exp, err_m_exp, m_theo, err_m_theo)
```

Wir verwenden den höchsten möglichen Wert des Gain-Bandwidth-Produkts (GBP), ohne Resonanzen zu erzeugen, und messen das Rauschspektrum im niedrigsten Frequenzbereich, der einen klaren Cut-off im SQUID-Leseausgang aufweist. Anhand dieses Rauschspektrums können wir die Temperatur des SQUIDs bestimmen, indem wir die temperatur- und frequenzabhängige spektrale Dichte fitten.

Die spektrale Dichte S(f) wird durch die folgende Gleichung beschrieben:

$$ S(f) = \dfrac{4 k_B T}{R} \dfrac{1}{1 + \frac{f^2}{f_c^2}}. $$

```python
x_53_a = readfiles2('5.3-noise-spectrum/5.3-noise-spectrum_1.csv')

# Konvertiere U von dBV in sqrtS_V [V / sqrt(Hz)]

Delta_f = np.abs(x_53_a[1, 0] - x_53_a[2,0])
sqrtS_V_53_a = 10 ** (x_53_a[:,1] / 20) / sqrt(Delta_f)

# Konvertiere sqrtS_V in sqrtS_I [A / sqrt(Hz)]

Amp = m_exp # V uA^-1

sqrtS_I_53_a = sqrtS_V_53_a / Amp # uA Hz^(-1/2)
sqrtS_I_53_a = sqrtS_I_53_a * 1e-6 # A Hz^(-1/2)
```

Wir können eine Kurve über die spektrale Dichte fitten, wobei wir die bekannten Werte für $k_b$ und $f_c$ verwenden, um die Temperatur $T$ unseres Systems zu bestimmen. Im Diagramm ist das Intervall markiert, das wir in unserer Anpassung berücksichtigt haben.

Zunächst können wir die kritische Frequenz $f_c$ berechnen, indem wir eine Eingangsspulen-Induktivität von $L=2,9 nH$ annehmen. Die kritische Frequenz lässt sich dann als
$$ f_c = \dfrac{R}{2 \pi L} $$
bestimmen.

```python
# Fit
def sqrt_SI(f, T, R):
    k_B = 1.380649e-23 # J K^-1
    L = 2.9e-9 # H
    f_c = R / (2 * pi * L)  
    
    return sqrt( (4 * k_B * T) / (R) * 1 / (1 + (f ** 2) / (f_c) ** 2) )


a = 10 # untere Grenze für Fit
b = 600 # obere Grenze für fit

popt, pcov = curve_fit(sqrt_SI, x_53_a[a:b, 0], sqrtS_I_53_a[a:b])
popt_T = popt[0]
pcov_T = sqrt(pcov[0, 0])
popt_R = popt[1] * 1e9
pcov_R = sqrt(pcov[1, 1]) * 1e9

plt.figure(figsize = (12, 7))

plt.loglog(x_53_a[:,0], sqrtS_I_53_a, color = 'dodgerblue', 
         label = '7.2 GHz')
plt.loglog(x_53_a[a:b,0], sqrt_SI(x_53_a[a:b, 0], *popt), label = 'Fit',
          color = 'tomato', linewidth = 3)
plt.xlim(1e-3,0.5e1)
plt.ylim(5e-8, 2e-7)
plt.title('FLL Spektrum bei 100 kOhm mit Fit', size = 16)
plt.xlabel('Frequenz [MHz]', size = 14)
plt.axvline(x_53_a[a, 0], ls = '--', color = 'gray')
plt.axvline(x_53_a[b, 0], ls = '--', color = 'gray', label = 'Interval für fit')
plt.text(1.4e-3, 8e-8, fitparameter('Temperatur: $T$ ', popt_T, pcov_T, 3, '[ K]'))


plt.ylabel('Spektraldichte $\sqrt{S_I}$ [A Hz$^{-1/2}$]', size = 14)
plt.legend(loc = 'best')


plt.show()
```

### Diskussion

Das der Widerstand beim Einzelstufen-SQUID so niedrig ist, ist ein sehr gutes Zeichen. Auch das die Widerstände generell sehr stabil waren, deutet darauf hin, dass der Restwiderstand in erster Linie von den verwendeten Kabeln und nicht von den SQUIDs selbst stammt und somit die Kabelressistanz als konstanter Parameter betrachtet werden kann.

Die Messung der Strom-Spannungs-Charakteristik ermöglichte die Bestimmung des Unterschieds zwischen den ganzzahligen und halbzahligen Flussquanten $\phi_0$. Die klare Differenzierung zwischen den Flussquanten ist ein Hinweis auf die korrekte Funktion des SQUIDs. 

Im FLL-Modus wurde der Verstärkungsfaktor $m_{exp}$ mit 0,68591 V/$\mu$A ermittelt, während der theoretische Wert $m_{theo}$ bei 0,495 V/$\mu$A lag. Die Abweichung zwischen experimentellem und theoretischem Verstärkungsfaktor ist signifikant, mit einem relativen Fehler von etwa 1,39%. Diese Diskrepanz ist groß und könnte auf mehrere Ursachen zurückzuführen sein, darunter systematische Fehler im Experiment, ungenaue Modellannahmen für die Berechnung des theoretischen Verstärkungsfaktors oder zusätzliche Verluste im System, die nicht im Modell berücksichtigt wurden. Besonders auffällig ist die hohe Standardabweichung von mehr als 3, was darauf hindeutet, dass die Messungen nicht ausreichend genau sind, um die theoretische Vorhersage gut zu reproduzieren. Dies könnte durch Rauschen oder Verzerrungen im Messsystem verursacht werden, die möglicherweise durch eine unzureichende Bandbreite des Verstärkers oder eine ungenaue Kalibrierung beeinflusst wurden.

Bei den Messungen des Zwei-Stufen-SQUIDs im Open-Loop-Modus ergaben sich Werte für Temperatur (T = 4.76 K) und Widerstand (R = 12.499 Ohm), die im Einklang mit der Temperatur von flüssigem Helium (etwa 4,2K) liegen und mit den zuvor gemessenen Werten für die Einzelstufen-SQUIDs stehen.

Es gibt jedoch einige Dinge, die man verbessern könnte: Es wäre hilfreich, einen tieferen Einblick in die Abhängigkeiten der zu messenden Werte vor dem Experiment zu erhalten, z.B. wenn etwas bei höheren Strömen besser funktioniert. Klarere Anweisungen in den relevanten Teilen des Skripts wären ebenfalls sehr hilfreich, besonders wenn es nicht immer eine erkennbare Beziehung zwischen den im Skript genannten Werten und den im theoretischen Hintergrund erwähnten Werten gibt.
