```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import pandas as pd
mpl.use("pgf")
mpl.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
%matplotlib inline
```

```python
%matplotlib inline
names = np.array(["1"])#, "C", "D", "E"])

def print_spectra(name):
    
    sample = pd.read_csv("C:/Users/notchok/Documents/5. Semester/FP/FP/F61/Protocol images/{}.txt".format(name),
                         sep = "\t", skiprows = 1)
    sample_tms = pd.read_csv("C:/Users/notchok/Documents/5. Semester/FP/FP/F61/Protocol images/{}p.txt".format(name),
                             sep = "\t", skiprows = 1)
    amplitude = sample.iloc[:, 1].values
    amplitude_tms = sample_tms.iloc[:,1].values
    f = sample.iloc[:,0].values
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# First plot
    ax1.plot(f, amplitude, label='Sample {}'.format(name))
    ax1.set_xlabel(r'Frequency  $f$ [Hz]')
    ax1.set_ylabel('Amplitude [u.a]')
    ax1.legend()

# Second plot
    ax2.plot(f, amplitude, label='Sample {} + TMS'.format(name), color='green')
    ax2.set_xlabel(r'Frequency  $f$ [Hz]')
    ax2.set_ylabel('Amplitude [u.a]')
    ax2.legend()
    
    #plt.savefig("freq_spectrum_sample{}.jpg".format(name))
    plt.show()
```

```python
print_spectra("A")
```
