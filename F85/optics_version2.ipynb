## Part 1: Polarization optics

How do we characterize the polarization of light?

Light can be described as an electromegnetic wave. Traveling wrt z as $E \left(z,t\right) = E_0 e^{i\left(\omega t - kz \right)}$ with $k = 2\pi / \lambda$. \\
The direction of oszillation of the electric field or the polarization vector $\vec{P}$ specifies the kind of polirization:
- Linear: Electric field oscillates along a straight line in the plane of polarization.
- Elliptic: Electric field describes an ellipse. If alligned with the coordinates one gets: $\vec{E}_\pm = E \left(a\hat{e}_x \mp ib\hat{e}_y\right)e^{i\left(\omega t - kx\right)}$
- Circular: Special case of elliptic, where the electric field describes a cyrcle.

The interaction of kinds of polarized light with each other is notable. 
Linear polarized light can be split up, or comprised of two other linear waves that are in phase $\vec{E} = \vec{E}_1 + \vec{E}_2$, or also be constructed by adding two counter-rotatingcircular polarized waves.

Similarly elliptical light can be a superposition of two linear polarized waves with a phase shift of $\pi /2$, where the amplitude of each wave determines the amplitude of the main axis. Thus for waves with equal amplitude one obtaines a circular wave.

How do we manipulate the polarization of light?

Waves plates: birefringetn crystals of a thickness of $\left(m + \frac{1}{4}\right)\lambda$ (quarter wave plate) or $\left(m + \frac{1}{2}\right)\lambda$ (half wave plate), where m is an integer. 

Half wave plate: Waves traveling along different optical axes experience a phase shift of $\pi$ relative to each other. Linear light would leave this plate again linear but rotated by twice the angle it entered the wave plate.

Quarter wave plate: Results in a phase shift of $\pi /2$. Resulting in a linear wave to convert into an elliptical wave or vice versa. For an incidence angle of $45^\circ$ one obtaines circular light.

How do we show that light is polarized?

To determine if light is polarized the light beam can be passed through a poralizer and then through a analyser. Depending on the change of intensity of the light one can determine if the light is poralized, partially polarized or unporalized. In our particular case we did this by using a half wave plate as analyzer and as poralizer.

### Abhängigkeit des Beugungswinkels von der Frequenz
Wir messen den Abstand zwischen unseren Maxima und erhalten den Winkel wie folgt:
$$\alpha=\arctan\left(\frac{d}{l}\right)$$
mit einem Fehler von:
 $$\Delta \alpha = \sqrt{\left(\frac{\partial \alpha}{\partial l}\Delta l\right)^2+\left(\frac{\partial \alpha}{\partial d}\Delta d\right)^2}= \sqrt{\left(\frac{d}{d^2+l^2}\Delta l\right)^2 + \left(\frac{l}{d^2+l^2}\Delta d\right)^2}$$


```python
l = 141  
delta_l = 2 

# first order
f0 = np.array([85,90,95,100,110,120,130,135]) 
delta_f0 = np.ones(8)*0.5 
dI = np.array([1.8,2.0,2.0,2.1,2.3,2.5,2.75,2.9])  
delta_dI = np.ones(8)*0.2 # [cm] 

# second order
f1 = np.array([85,90,95,100,110,120])  
delta_f1 = np.ones(6)*0.5 
dII = np.array([2.6,3.8,4,4.3,4.65,5.1]) 
delta_dII = np.ones(6)*0.2  

# calculate the angles
def alpha(l,d):
    return np.arctan(d/l)*180/(np.pi)
def delta_alpha(l,d,delta_l,delta_d):
    return np.sqrt( (d/(d**2+l**2)*delta_l)**2 + (l/(d**2+l**2)*delta_d)**2 ) *180/(np.pi)

alpha_I = alpha(l,dI)
delta_alpha_I = delta_alpha(l,dI,delta_l,delta_dI)

alpha_II = alpha(l,dII)
delta_alpha_II = delta_alpha(l,dII,delta_l,delta_dII)

poptI, pcovI = curve_fit(linear_model, f0, alpha_I, sigma=delta_alpha_I)
poptII, pcovII = curve_fit(linear_model, f1, alpha_II, sigma=delta_alpha_II)
plt.errorbar(f0, alpha_I, xerr=delta_f0, yerr=delta_alpha_I, fmt='x')
plt.plot(f0, linear_model(f0, *poptI), label='1st maximum')

plt.errorbar(f1, alpha_II, xerr=delta_f1, yerr=delta_alpha_II, fmt='x')
plt.plot(f1, linear_model(f1, *poptII), label='2nd maximum')

plt.title("Winkel zum n'ten Maximum gegen Frequenz")
plt.xlabel("Frequenz  [MHz]")
plt.ylabel("Winkel  [°]")
plt.grid()
plt.legend()
plt.show()
```

```python
# fit coefficients:
print('a_I = (', np.round(poptI[0],4), "+/-", np.round(np.sqrt(pcovI[0][0]),4), '°/MHz')
print('b_I = (', np.round(poptI[1],2), "+/-", np.round(np.sqrt(pcovI[1][1]),2), '°')
print('a_II = (', np.round(poptII[0],3), "+/-", np.round(np.sqrt(pcovII[0][0]),3), '°/MHz')
print('b_II = (', np.round(poptII[1],1), "+/-", np.round(np.sqrt(pcovII[1][1]),1), '°')
```

### Berechnung der Schallgeschwindigkeit
Mit hilfe von $\sin\theta_B=\frac{1}{2}\frac{m\lambda}{\lambda_s}$ und $v=\lambda f$ finden wir für kleine Winkel:

$$ \sin\theta_B\frac{v_s}{f_s}=\frac{1}{2}m\lambda$$
$$ \Rightarrow \quad v_s = \frac{1}{2}m\lambda\frac{f_s}{\sin\theta_B}\approx\frac{1}{2}m\lambda\frac{f_s}{\theta_B}=\frac{1}{2}m\lambda a$$

$$ \Delta v_s = \frac{1}{2}m\lambda \Delta a = v_s\frac{\Delta a}{a}$$

Mit einer Wellenänge von $\lambda \approx 633nm$ erhalten wir unter Betrachtung des ersten Maxima folgende Schallgeschwindigkeit:


```python
lambd = 633*10**(-9) 
a = poptI[0]*np.pi/180*1e-6  
delta_a = np.sqrt(pcovI[0][0])*np.pi/180*1e-6  

v_s = lambd/(2*a)
delta_v_s = v_s*delta_a/a

print('v_s = ', np.round(v_s,0), '+/-', np.round(delta_v_s,0), ' m/s')
```

### Berechnung der Wellenlänge und Gitterperiode
Mit dieser Schallgeschwindigkeit können wir die Wellenlänge berechnen:
$$\lambda_s=\frac{v_s}{f_s}$$
$$ \Delta \lambda_s = \lambda_s\sqrt{\left(\frac{\Delta v_s}{v_s}\right)^2+\left(\frac{\Delta f_s}{f_s}\right)^2}$$

```python
f3= f0*10**6   
delta_f3= delta_f0*10**6

av_lambd_s = np.mean(v_s/(f3))
delta_av_lambd_s = np.mean(v_s/f3*np.sqrt((delta_v_s/v_s)**2 + (delta_f3/f3)**2))

print('lambda_s = ', np.round(av_lambd_s,7), '+/-', np.round(delta_av_lambd_s,7), 'm')
```

Wir erhalten eine Wellenlänge von $\lambda_s = (20.4\pm0.9)\mu m$.

Mit dem $L$ aus dem Skript können wir nun die Gitterkonstante berechnen:
$$n=\frac{L}{\lambda_s}$$
$$  \Delta n = n\frac{\Delta \lambda_s}{\lambda_s}$$

```python
L = 20 

n = L*10**(-3)/(av_lambd_s) 
delta_n = n*delta_av_lambd_s/av_lambd_s

print('n =', np.round(n,0), '+-', np.round(delta_n,0))
```

### Berechnung der Schallleistung bei optimaler Ablenkungseffizienz.
Für die Schallleistung können wir die Gleichung (3.4) aus dem Skript nutzen.
$$ R= \sin^2\left(\frac{\pi L}{\lambda \cos \theta_B}\sqrt{\frac{M_2 I_s}{2}}\right) $$
Bei einer optimalen Ablenkungseffizienz gilt $R=1$ und $\cos(\theta_B)=1$, sodass 
$$\frac{\pi}{2}=\frac{\pi L}{\lambda \cos \theta_B}\sqrt{\frac{M_2 I_s}{2}}$$
$$\Rightarrow I_s = \dfrac{\lambda^2}{2\,L^2\, M_2}$$ 
Die Schallleistung $P$ ergibt sich aus $I_s$ und der Fläche $A$
$$P=A I_s = L h I_s = L  h \, \dfrac{\lambda^2}{2\,L^2\, M_2} = \dfrac{h\lambda^2}{2\,L\, M_2} $$

```python

h = 3e-3        
M_2 = 34.5e-15  

P_s = h*lambd**2/(2*(L*10**(-3))*M_2)

print('P_s =', P_s, 'W')
```
