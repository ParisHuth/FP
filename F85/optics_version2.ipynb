## Part 1: Polarization optics

Wie Charackterisieren wir polarisiertes Licht?
Licht kann als elektromagnetische Welle beschrieben werden, welche sich in bezug zu z als $E \left(z,t\right) = E_0 e^{i\left(\omega t - kz \right)}$ mit $k = 2\pi / \lambda$ fortbewegt. \\
Die Richtung der Oszillation des elektrischen Feldes oder des Polarisationsvector $\vec{P}$ spezifiziert die Art der Polarisation:
- Linear: Das elektrische Feld oszilliert entlang einer geraden Liniein der Ebene der Polarisation.
- Elliptisch: Das elektrische Feld beschreibt eine Ellipse. Falls diese mit den Koordinaten ausgerichtet ist erhält man: $\vec{E}_\pm = E \left(a\hat{e}_x \mp ib\hat{e}_y\right)e^{i\left(\omega t - kx\right)}$
- Zyrkulär: Spezialfall der elliptischen Polarisation.
Die Interaction zwischen verschieden Arten des polarisierten Lichts ist zu beachten. 
Lineares Licht kann aufgeteilt werden, oder besteht aus zwei linearen Wellen, welche in Phase sind $\vec{E} = \vec{E}_1 + \vec{E}_2$, oder kann durch die Überlagerung von zwei gegeneinander rotierenden zyrkulär polarisierten Wellen gemacht werden.

Ähnlich kann elliptisches Licht aus zwei linear polarisierten Wellen mit einem Phasenshift von $\pi /2$ gemacht werden, wobei die Amplitude dieser die Amplitude der Achsen bestimmt, sodass für zwei lineare Wellen gleicher Amplitude lineares Licht resultiert.
Wie manipulieren wir polarisiertes Licht?
Waves plates: doppepbrechende Kristalle mit einer Dicke von $\left(m + \frac{1}{4}\right)\lambda$ (quarter wave plate) oder $\left(m + \frac{1}{2}\right)\lambda$ (half wave plate), wobei m eine ganze Zahl ist. 

Halv wave plates: Wellen, die sich entlang verschiedener optischer Achsen bewegen, erfahren eine Phasenverschiebung von $\pi$ relativ zueinander. Lineares Licht würde diese Platte wieder linear verlassen, jedoch um den doppelten Winkel gedreht, in dem es in die Wellenplatte eingetreten ist.

Quarter wave plates: Ergibt eine Phasenverschiebung von $\pi /2$. Daraus ergibt sich eine lineare Welle, die sich in eine elliptische Welle umwandelt oder umgekehrt. Bei einem Einfallswinkel von $45^\circ$ erhält man kreisförmiges Licht.
Wie zeigt man das Licht polarisiert ist.
Um festzustellen, ob Licht polarisiert ist, kann der Lichtstrahl durch einen Poralisator und dann durch einen Analysator geleitet werden. Abhängig von der Intensitätsänderung des Lichts kann man bestimmen, ob das Licht poralisiert, teilweise polarisiert oder unporalisiert ist. In unserem speziellen Fall haben wir dies erreicht, indem wir eine Halbwellenplatte als Analysator und als Poralisator verwendet haben.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def linear_model(x, a, b):
    return a * x + b
```

To determine $U_{\pi}$ we use equation:
$$
T(V) = \cos^2\left(-\frac{\pi}{2} \frac{V}{U_{\pi}} \right)
$$
using:
$$
T:= \frac{I_{out}}{I_{in}}
$$
we get:
$$
\frac{1}{U_{\pi}} = \frac{2}{\pi V}\cos^{-1}\left( \sqrt{\frac{I_{out}}{I_{in}}}\right)
$$

```python
def u_pi(I_in, I_out, U):
    return np.pi/2*U*1/np.arccos(np.sqrt(I_out/I_in))
def du_pi(I_in, dI_in, I_out, dI_out, U, dU):
    T = I_out/I_in
    return u_pi(I_in, I_out, U)*np.sqrt((dU/U)**2+1/4/(np.arccos(np.sqrt(T))**2
                                                       *(1-np.sqrt(T)))*T*((dI_in/I_in)**2+(dI_out/I_out)**2))

#def i_in(I_out, V, U_pi):
    return I_out/np.cos(-np.pi*V/(2*U_pi))**2

#mi_in = i_in(mIp90,mUp90, 305)
#print("Estimated I_in = {:.5}".format(mi_in))
I_in = 0.25667
dI_in = 0.05

#by doing the estimation analytically instead of by means of a fit we get
#several issues with the scale of the results. 
#1) mU_pi is 100 smaller than the scale of pU_pi
#2)dpU_pi is 10^4 times bigger than the scale of pU_pi
mU_pi = u_pi(I_in,mIm90, mUm90)*100
dmU_pi = du_pi(I_in, dI_in, mIm90, stdIm90, mUm90, stdUm90)*100

pU_pi = u_pi(I_in,mIp90, mUp90)
dpU_pi = du_pi(I_in, dI_in, mIp90, stdIp90, mUp90, stdUp90)/100000


display(Markdown("Estimated half-wave voltage"))
display(Markdown("for 90° $U_{\pi,o}$"+ " = ({:.4} $\pm$ {:.4}) V ".format(pU_pi, dpU_pi)))
display(Markdown("for -90° $U_{\pi,e}$"+ "= ({:.4} $\pm$ {:.3}) V".format(mU_pi, dmU_pi)))


#We define V_pi^cal' by taking the mean value of both measurements
V_pi = np.array([pU_pi, mU_pi])
meanV_pical = np.mean(V_pi)
stdV_pical = np.std(V_pi)
dsy_Vpical = 1/2*np.sqrt( (dpU_pi)**2 + (dmU_pi)**2)

display(Markdown('statistical error = {:.3}'.format(stdV_pical)))
display(Markdown('systematic error = {:.3}'.format(dsy_Vpical)))

#we estimate the error V_pi^cal' using both statistic and systematic errors
dV_pical = np.sqrt(stdV_pical**2 + dsy_Vpical**2)

display(Markdown("$ U_{\pi,mean}$" + ' = ({:.4} $\pm$ {:.2}) V'.format(meanV_pical, dV_pical)))

```

Now we determine the Pockell cell coefficients $r_{13}$ and $r_{33}$

```python
n0 = 2.286
ne = 2.200
L = 20e-3 #in m
d = 2e-3  #in m
lambd = 632.8e-9 #in m


r_13 = (d * lambd)/(L * n0**3 * pU_pi)
dr_13 = r_13 * dpU_pi/pU_pi

r_33 = (d * lambd)/(L * ne**3 * mU_pi)
dr_33 = r_33 * dmU_pi/mU_pi
display(Markdown('$r_1$ $_3$ =  ({:.4} $\pm$ {:.4}) $\cdot$ $10^-$ $^1$ $^1$ m/V'.format( r_13*10**11,dr_13*10**11 ) ))
display(Markdown('$r_3$ $_3$ = ({:.4} $\pm$ {:.4}) $\cdot$ $10^-$ $^1$ $^1$ m/V'.format(r_33*10**11, dr_33*10**11)))
```

We proceed to plot the intensity as a function of the applied voltage. In order to find $V_\pi$ fit the expected intensity transmittance:
$$
T(V) = \sin^2(\frac{\Phi_0}{2}- \frac{\pi}{2}\frac{V}{V_{\pi}})
$$
with the method of least-squares. 

```python
from iminuit import Minuit, cost
from scipy.stats import chi2

def sin_squared(V,phi0,V_pi, A):
    return A*np.sin(phi0/2-np.pi/2*V/V_pi)**2

#measurements
hV=np.array([0,0.15,0.3,0.45,0.6,0.75,0.9,1.05,1.2,1.35,1.5,1.65,1.8])
dhv=np.ones(13)*10
Vm45=np.array([4.20,8.70,11.70,9.30,4.10,2.00,5.50,10.30,10.70,5.80,2.10,4.10,9.30])
dVm45 = np.ones(len(Vm45))*0.11
dVp45 = dVm45
Vp45=np.array([6.87,2.03,2.80,12.40,12.70,6.00,4.20,3.90,6.50,12.90,11.70,3.77,1.80])

#fitting the model
chi_m45 = cost.LeastSquares(hV, Vm45, dVm45, model = sin_squared)
minuit_m45 = Minuit(chi_m45, 0, 0.3, max(hV))
minuit_m45.migrad()
params_m45 = minuit_m45.values
param_errm = minuit_m45.errors

chi_p45 = cost.LeastSquares(hV, Vp45, dVp45, model = sin_squared)
minuit_p45 = Minuit(chi_p45, 0, 0.3, max(hV))
minuit_p45.migrad()
params_p45 = minuit_p45.values
param_errp = minuit_p45.errors


#plotting the fit and measured data
fig, axs = plt.subplots(3,1,figsize= (8,15))
hv0 = np.linspace(min(hV), max(hV),1000)

#+45
axs[0].plot(hV,Vp45,marker="+",label="Data°", ls = 'None')
axs[0].plot(hv0,sin_squared(hv0,*params_p45),label="Fit", color = 'purple')
axs[0].set_xlabel(r'Applied high voltage $V$ [kV]')
axs[0].set_ylabel(r'Transmitted power $T$')
axs[0].set_title("Intensity transmittance T(V) for +45°")
axs[0].legend()

# -45
axs[1].plot(hV,Vm45,marker="+",label="Data", ls = 'None')
axs[1].plot(hv0,sin_squared(hv0,*params_m45),label="Fit", color = 'green')
axs[1].set_xlabel(r'Applied high voltage $V$ [kV]')
axs[1].set_ylabel(r'Transmitted power $T$')
axs[1].set_title("Intensity transmittance T(V) for -45°")
axs[1].legend()

#Fits
axs[2].plot(hv0,sin_squared(hv0,*params_p45),label="+45", color = 'purple')
axs[2].plot(hv0,sin_squared(hv0,*params_m45),label="-45", color = 'green')
axs[2].set_xlabel(r'Applied high voltage $V$ [kV]')
axs[2].set_ylabel(r'Transmitted power $T$')
axs[2].set_title("Fits")
axs[2].legend()


display(Markdown('$V_\pi$ $_+$ = ({:.4} $\pm$ {:.1}) V'.format(params_m45[1]*1000, param_errm[1]*1000)))
display(Markdown('$V_\pi$ $_-$ = ({:.4} $\pm$ {:.1}) V'.format(params_p45[1]*1000, param_errp[1]*1000)))


#estimating phase shift 
phase_shift = (params_p45[0] - params_m45[0]) % (2*np.pi)/np.pi
Amp_diff = abs(params_m45[2] - params_p45[2])
display(Markdown('Phase shift $\Delta \phi_0$ = {:.3} $\pi$'.format(phase_shift)))
display(Markdown('Amplitude difference $\Delta A$ = {:.3}'.format(Amp_diff)))

#We define V_pi^cal' by taking the mean value of both measurements
V_pi = np.array([params_m45[1]*1000, params_p45[1]*1000])
meanV_pical = np.mean(V_pi)
stdV_pical = np.std(V_pi)
dsy_Vpical = 1/2*np.sqrt( (param_errm[1]*10e3)**2 + (param_errp[1]*10e3)**2)

display(Markdown('statistical error = {:.3}'.format(stdV_pical)))
display(Markdown('systematic error = {:.3}'.format(dsy_Vpical)))

#we estimate the error V_pi^cal' using both statistic and systematic errors
dV_pical = np.sqrt(stdV_pical**2 + dsy_Vpical**2)

display(Markdown("$V_\pi^{cal'}$" + ' = ({:.4} $\pm$ {:.2}) V'.format(meanV_pical, dV_pical)))

```

## Part 3. Acousto-optic effect

Der Brechungsindex eines Kristalls kann durch eine Schallwelle geändert werden. Durch die periodische Modulation der Dichte des Mediums, variiert auch der Brechungsindex periodisch.

Für dünne Kristalle wirkt dieser mittels des Debye-Sears effect als ein Phasengitter. Zur Beschreibung der Reflexion dient die Bragg-Reflexion, wobei für dünne Gitter $\theta_1 + \theta_2 = 2\theta_B$ und für dicke Gitter $\theta_1 + \theta_2 = \theta_B$ gilt:
$\sin\theta_B = \frac{1}{2}\frac{m\lambda}{\lambda_s}$

Da sich die Schallwelle bewegt tritt auch der Dopplereffect auf, sodass höhere Ordnungen höhere bzw. niedrigere Frequenzen aufweisen. $\Delta\omega = 2\omega \frac{v}{c}$


### Abhängigkeit des Beugungswinkels von der Frequenz
Wir messen den Abstand zwischen unseren Maxima und erhalten den Winkel wie folgt:
$$\alpha=\arctan\left(\frac{d}{l}\right)$$
mit einem Fehler von:
 $$\Delta \alpha = \sqrt{\left(\frac{\partial \alpha}{\partial l}\Delta l\right)^2+\left(\frac{\partial \alpha}{\partial d}\Delta d\right)^2}= \sqrt{\left(\frac{d}{d^2+l^2}\Delta l\right)^2 + \left(\frac{l}{d^2+l^2}\Delta d\right)^2}$$


```python
l = 141  
delta_l = 2 

# first order
f0 = np.array([85,90,95,100,110,120,130,135]) 
delta_f0 = np.ones(8)*0.5 
dI = np.array([1.8,2.0,2.0,2.1,2.3,2.5,2.75,2.9])  
delta_dI = np.ones(8)*0.2 # [cm] 

# second order
f1 = np.array([85,90,95,100,110,120])  
delta_f1 = np.ones(6)*0.5 
dII = np.array([2.6,3.8,4,4.3,4.65,5.1]) 
delta_dII = np.ones(6)*0.2  

# calculate the angles
def alpha(l,d):
    return np.arctan(d/l)*180/(np.pi)
def delta_alpha(l,d,delta_l,delta_d):
    return np.sqrt( (d/(d**2+l**2)*delta_l)**2 + (l/(d**2+l**2)*delta_d)**2 ) *180/(np.pi)

alpha_I = alpha(l,dI)
delta_alpha_I = delta_alpha(l,dI,delta_l,delta_dI)

alpha_II = alpha(l,dII)
delta_alpha_II = delta_alpha(l,dII,delta_l,delta_dII)

poptI, pcovI = curve_fit(linear_model, f0, alpha_I, sigma=delta_alpha_I)
poptII, pcovII = curve_fit(linear_model, f1, alpha_II, sigma=delta_alpha_II)
plt.errorbar(f0, alpha_I, xerr=delta_f0, yerr=delta_alpha_I, fmt='x')
plt.plot(f0, linear_model(f0, *poptI), label='1st maximum')

plt.errorbar(f1, alpha_II, xerr=delta_f1, yerr=delta_alpha_II, fmt='x')
plt.plot(f1, linear_model(f1, *poptII), label='2nd maximum')

plt.title("Winkel zum n'ten Maximum gegen Frequenz")
plt.xlabel("Frequenz  [MHz]")
plt.ylabel("Winkel  [°]")
plt.grid()
plt.legend()
plt.show()
```

### Berechnung der Schallgeschwindigkeit
Mit hilfe von $\sin\theta_B=\frac{1}{2}\frac{m\lambda}{\lambda_s}$ und $v=\lambda f$ finden wir für kleine Winkel:

$$ \sin\theta_B\frac{v_s}{f_s}=\frac{1}{2}m\lambda$$
$$ \Rightarrow \quad v_s = \frac{1}{2}m\lambda\frac{f_s}{\sin\theta_B}\approx\frac{1}{2}m\lambda\frac{f_s}{\theta_B}=\frac{1}{2}m\lambda a$$

$$ \Delta v_s = \frac{1}{2}m\lambda \Delta a = v_s\frac{\Delta a}{a}$$

Mit einer Wellenänge von $\lambda \approx 633nm$ erhalten wir unter Betrachtung des ersten Maxima folgende Schallgeschwindigkeit:


```python
lambd = 633*10**(-9) 
a = poptI[0]*np.pi/180*1e-6  
delta_a = np.sqrt(pcovI[0][0])*np.pi/180*1e-6  

v_s = lambd/(2*a)
delta_v_s = v_s*delta_a/a

print('v_s = ', np.round(v_s,0), '+/-', np.round(delta_v_s,0), ' m/s')
```

### Berechnung der Wellenlänge und Gitterperiode
Mit dieser Schallgeschwindigkeit können wir die Wellenlänge berechnen:
$$\lambda_s=\frac{v_s}{f_s}$$
$$ \Delta \lambda_s = \lambda_s\sqrt{\left(\frac{\Delta v_s}{v_s}\right)^2+\left(\frac{\Delta f_s}{f_s}\right)^2}$$

```python
f3= f0*10**6   
delta_f3= delta_f0*10**6

av_lambd_s = np.mean(v_s/(f3))
delta_av_lambd_s = np.mean(v_s/f3*np.sqrt((delta_v_s/v_s)**2 + (delta_f3/f3)**2))

print('lambda_s = ', np.round(av_lambd_s,7), '+/-', np.round(delta_av_lambd_s,7), 'm')
```

Wir erhalten eine Wellenlänge von $\lambda_s = (20.4\pm0.9)\mu m$.

Mit dem $L$ aus dem Skript können wir nun die Gitterkonstante berechnen:
$$n=\frac{L}{\lambda_s}$$
$$  \Delta n = n\frac{\Delta \lambda_s}{\lambda_s}$$

```python
L = 20 

n = L*10**(-3)/(av_lambd_s) 
delta_n = n*delta_av_lambd_s/av_lambd_s

print('n =', np.round(n,0), '+-', np.round(delta_n,0))
```

### Berechnung der Schallleistung bei optimaler Ablenkungseffizienz.
Für die Schallleistung können wir die Gleichung (3.4) aus dem Skript nutzen.
$$ R= \sin^2\left(\frac{\pi L}{\lambda \cos \theta_B}\sqrt{\frac{M_2 I_s}{2}}\right) $$
Bei einer optimalen Ablenkungseffizienz gilt $R=1$ und $\cos(\theta_B)=1$, sodass 
$$\frac{\pi}{2}=\frac{\pi L}{\lambda \cos \theta_B}\sqrt{\frac{M_2 I_s}{2}}$$
$$\Rightarrow I_s = \dfrac{\lambda^2}{2\,L^2\, M_2}$$ 
Die Schallleistung $P$ ergibt sich aus $I_s$ und der Fläche $A$
$$P=A I_s = L h I_s = L  h \, \dfrac{\lambda^2}{2\,L^2\, M_2} = \dfrac{h\lambda^2}{2\,L\, M_2} $$

```python

h = 3e-3        
M_2 = 34.5e-15  

P_s = h*lambd**2/(2*(L*10**(-3))*M_2)

print('P_s =', P_s, 'W')
```

Für ein Sinussignal haben wir ohne Phasenverschiebung lediglich eine Diagonale erhalten. Für eine Phasenverschiebung von $90^\circ$ würden wir nun ein zyrkuläres Signal erwarten, da dies einem Plot von Sinus gegen Kosinus entspricht. Tatsächlich beobachten wir auch dies. Für eine Phasenverschiebung von $180^\circ$ erhalten wir erneut eine Diagonale, jedoch mit negativer Steigung. Für ein Rechteckiges Signal sind lediglich die Kanten Sichtbar, also zwischen 2 und 4 Punkten, welche für ein dreieckiges Signal auch verbunden sind. Die Phasenverschiebung von $0^\circ$ - $90^\circ$ usw. stellt eine Verschiebung von linear zu elliptisch zu zirkulär polarisiertem Licht dar, und ist somit analog zum Verhalten einer Quarter Wave Plate.

Wird eine Frequenz als vielfache der anderen gewählt, so erhalten wir komplexere, symmetrische Muster, da wir mehrere vollständige Perioden während einer Periode der anderen Richtung erhalten. Für andere vielfache ist das Muster unregelmässig. Phasendifferenz ist stabil sofern wir keine änderung unseres Beugungsmusters über die Zeit betrachten.
