## Part 1: Polarization optics

How do we characterize the polarization of light?

Light can be described as an electromegnetic wave. Traveling wrt z as $E \left(z,t\right) = E_0 e^{i\left(\omega t - kz \right)}$ with $k = 2\pi / \lambda$. \\
The direction of oszillation of the electric field or the polarization vector $\vec{P}$ specifies the kind of polirization:
- Linear: Electric field oscillates along a straight line in the plane of polarization.
- Elliptic: Electric field describes an ellipse. If alligned with the coordinates one gets: $\vec{E}_\pm = E \left(a\hat{e}_x \mp ib\hat{e}_y\right)e^{i\left(\omega t - kx\right)}$
- Circular: Special case of elliptic, where the electric field describes a cyrcle.

The interaction of kinds of polarized light with each other is notable. 
Linear polarized light can be split up, or comprised of two other linear waves that are in phase $\vec{E} = \vec{E}_1 + \vec{E}_2$, or also be constructed by adding two counter-rotatingcircular polarized waves.

Similarly elliptical light can be a superposition of two linear polarized waves with a phase shift of $\pi /2$, where the amplitude of each wave determines the amplitude of the main axis. Thus for waves with equal amplitude one obtaines a circular wave.

How do we manipulate the polarization of light?

Waves plates: birefringetn crystals of a thickness of $\left(m + \frac{1}{4}\right)\lambda$ (quarter wave plate) or $\left(m + \frac{1}{2}\right)\lambda$ (half wave plate), where m is an integer. 

Half wave plate: Waves traveling along different optical axes experience a phase shift of $\pi$ relative to each other. Linear light would leave this plate again linear but rotated by twice the angle it entered the wave plate.

Quarter wave plate: Results in a phase shift of $\pi /2$. Resulting in a linear wave to convert into an elliptical wave or vice versa. For an incidence angle of $45^\circ$ one obtaines circular light.

How do we show that light is polarized?

To determine if light is polarized the light beam can be passed through a poralizer and then through a analyser. Depending on the change of intensity of the light one can determine if the light is poralized, partially polarized or unporalized. In our particular case we did this by using a half wave plate as analyzer and as poralizer.

To determine $U_{\pi}$ we use equation:
$$
T(V) = \cos^2\left(-\frac{\pi}{2} \frac{V}{U_{\pi}} \right)
$$
using:
$$
T:= \frac{I_{out}}{I_{in}}
$$
we get:
$$
U_{\pi} =\frac{\pi V}{2}\arccos^{-1}\left( \sqrt{\frac{I_{out}}{I_{in}}}\right)
$$

```python
def u_pi(I_in, I_out, U):
    return np.pi/2*U*1/np.arccos(np.sqrt(I_out/I_in))
def du_pi(I_in, dI_in, I_out, dI_out, U, dU):
    T = I_out/I_in
    return u_pi(I_in, I_out, U)*np.sqrt((dU/U)**2+1/4/(np.arccos(np.sqrt(T))**2
                                                       *(1-np.sqrt(T)))*T*((dI_in/I_in)**2+(dI_out/I_out)**2))

#def i_in(I_out, V, U_pi):
    return I_out/np.cos(-np.pi*V/(2*U_pi))**2

#mi_in = i_in(mIp90,mUp90, 305)
#print("Estimated I_in = {:.5}".format(mi_in))
I_in = 0.25667
dI_in = 0.05

#by doing the estimation analytically instead of by means of a fit we get
#several issues with the scale of the results. 
#1) mU_pi is 100 smaller than the scale of pU_pi
#2)dpU_pi is 10^4 times bigger than the scale of pU_pi
mU_pi = u_pi(I_in,mIm90, mUm90)*100
dmU_pi = du_pi(I_in, dI_in, mIm90, stdIm90, mUm90, stdUm90)*100

pU_pi = u_pi(I_in,mIp90, mUp90)
dpU_pi = du_pi(I_in, dI_in, mIp90, stdIp90, mUp90, stdUp90)/100000


display(Markdown("Estimated half-wave voltage"))
display(Markdown("for 90° $U_{\pi,o}$"+ " = ({:.4} $\pm$ {:.4}) V ".format(pU_pi, dpU_pi)))
display(Markdown("for -90° $U_{\pi,e}$"+ "= ({:.4} $\pm$ {:.3}) V".format(mU_pi, dmU_pi)))


#We define V_pi^cal' by taking the mean value of both measurements
V_pi = np.array([pU_pi, mU_pi])
meanV_pical = np.mean(V_pi)
stdV_pical = np.std(V_pi)
dsy_Vpical = 1/2*np.sqrt( (dpU_pi)**2 + (dmU_pi)**2)

display(Markdown('statistical error = {:.3}'.format(stdV_pical)))
display(Markdown('systematic error = {:.3}'.format(dsy_Vpical)))

#we estimate the error V_pi^cal' using both statistic and systematic errors
dV_pical = np.sqrt(stdV_pical**2 + dsy_Vpical**2)

display(Markdown("$ U_{\pi,mean}$" + ' = ({:.4} $\pm$ {:.2}) V'.format(meanV_pical, dV_pical)))

```

Now we determine the Pockell cell coefficients $r_{13}$ and $r_{33}$

```python
n0 = 2.286
ne = 2.200
L = 20e-3 #in m
d = 2e-3  #in m
lambd = 632.8e-9 #in m


r_13 = (d * lambd)/(L * n0**3 * pU_pi)
dr_13 = r_13 * dpU_pi/pU_pi

r_33 = (d * lambd)/(L * ne**3 * mU_pi)
dr_33 = r_33 * dmU_pi/mU_pi
display(Markdown('$r_1$ $_3$ =  ({:.4} $\pm$ {:.4}) $\cdot$ $10^-$ $^1$ $^1$ m/V'.format( r_13*10**11,dr_13*10**11 ) ))
display(Markdown('$r_3$ $_3$ = ({:.4} $\pm$ {:.4}) $\cdot$ $10^-$ $^1$ $^1$ m/V'.format(r_33*10**11, dr_33*10**11)))
```

Für eine Phasenverschiebung von $90^\circ$ würden wir nun ein zyrkuläres Signal erwarten, da dies ein Plot von Sinugs gegen Kosinus entspricht. Tatsächlich beobachten wir auch dies.
Für eine Phasenverschiebung von $180^\circ$ erhalten wir erneut eine Diagonale, jedoch mit negativer Steigung. Die Manipulation von unpolarisertem Licht durch die beiden AOM ist mit der manipulation durch Polifilter und Wave Plates zu vergleichen, wobei hier Brechungsbilder durch die Modulation von Brechungsindexen und nicht die gezielte Behandlung von Polirisiertem Licht geschieht. Die Phasenverschiebung von $0^\circ - 90^\circ$ usw. stellt eine verschiebung von linear zu elliptisch zu zyrkulär polarisiertem Licht dar, und ist somit analog zum verhalten einer Quarter Wave Plate.

Wird eine Frequenz als vielfache der Anderen gewählt, so erhalten wir komplexere, symmetrische Muster. Da wir meherer vollständige Perioden während einer Periode der anderen Richtung erhalten. Für ungerade vielfache ist das Muster unregelmäßig. Phasendifferenz ist stabil sofern wir keine Änderung unseres Beugungsmusters über Zeit betrachten.
